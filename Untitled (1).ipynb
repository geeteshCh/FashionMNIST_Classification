```python
import matplotlib.pyplot as plt
import numpy as np
from keras.datasets import fashion_mnist
# load dataset
(trainX, trainy), (testX, testy) = fashion_mnist.load_data()
```

```python
# summarize loaded dataset
print('Train: X=%s, y=%s' % (trainX.shape, trainy.shape))
print('Test: X=%s, y=%s' % (testX.shape, testy.shape))
```

```python
 # reshape dataset to have a single channel
 trainX = trainX.reshape((trainX.shape[0], 28, 28, 1))
 testX = testX.reshape((testX.shape[0], 28, 28, 1))
```

```python
# convert from integers to floats
trainX_norm = trainX.astype('float32')
testX_norm = testX.astype('float32')
# normalize to range 0-1
trainX_norm = trainX_norm / 255.0
testX_norm = testX_norm / 255.0
```

```python
trainy_norm = trainy.astype('float32')
testy_norm = testy.astype('float32')
# normalize to range 0-1
trainy_norm = trainy_norm / 255.0
testy_norm = testy_norm / 255.0
```

```python
from __future__ import print_function
import time
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn import decomposition
%matplotlib inline
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns
```

```python
reshaped_data = trainX.reshape(-1, trainX.shape[0])
pca_dt = reshaped_data
# Set first three principle components
pca = PCA(n_components=0.9)
pca.fit(pca_dt)
# Fit the model with pixel columns and apply the dimensionality reduction on those columns
pca_result = pca.transform(pca_dt)
```

```python
# initializing the pca
# Set n_components as 784 dimensions
pca_all = decomposition.PCA(n_components = 784)
# Fit on training data
pca_784 = pca_all.fit_transform(pca_dt)
```

```python
print("Original Data Shape:\n", trainX.shape)
print("\nReshaped Data (2D):\n", reshaped_data.shape)
```

```python
# Inverse transform to approximate the original data
approx_original_data = pca.inverse_transform(pca_result)

# Reshape the approximated original data to the original shape
approx_original_data = approx_original_data.reshape(trainX.shape)
print(pca_result.shape)
```
